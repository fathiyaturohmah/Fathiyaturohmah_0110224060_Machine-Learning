Mengubah tipe data kolom MG

```python
df_copy = df.copy()
df_copy['Mg'] = pd.to_numeric(df_copy['Mg'], errors='coerce')
df_copy = df_copy.dropna()
df = df_copy
```

Memilih fitur dgn bbrapa kolom

```python
X_3 = df[['b2', 'b3', 'b4', 'b8', 'b11', 'Sigma_VV', 'Sigma_VH']]
y_3 = df['N']
```

Splitting data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_3, y_3, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

#training data
model.fit(X_train, y_train)
```

Evaluasi data ( Menggunakan bbrapa fitur kolom )

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

#testing
y_pred = model.predict(X_test)

#evaluasi
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("R2 Score:", r2)
print("RMSE:", mse)
```

Menggunakan semua fitur kolom

```python
#testing
y_pred = model.predict(X_test)

#evaluasi
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("R2 Score:", r2)
print("RMSE:", mse)
```

```python
coeff = pd.DataFrame({
    'Fitur': X_3.columns,
    'Koefisien': model.coef_
})

print(coeff)
```

```python
import statsmodels.api as sm

X_sm = sm.add_constant(X_3)
model_ols = sm.OLS(y_3, X_sm).fit()
print(model_ols.summary())
```
